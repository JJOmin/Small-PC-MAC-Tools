```python
import os
import time
import threading
import tkinter as tk
from tkinter import messagebox

shutdown_process = None

def update_timer():
    global shutdown_process
    
    while shutdown_process is not None:
        remaining_time = shutdown_process - time.time()
        if remaining_time <= 0:
            break
        minutes, seconds = divmod(int(remaining_time), 60)
        timer_label.config(text=f"Time left: {minutes:02d}:{seconds:02d}")
        time.sleep(1)
    timer_label.config(text="Time left: 00:00")
    root.deiconify()  # Show the main window again
    start_button.config(state=tk.NORMAL)  # Re-enable the start button

def shutdown_timer():
    global shutdown_process
    
    try:
        minutes = int(minutes_entry.get())
        seconds = int(seconds_entry.get())
        
        total_seconds = minutes * 60 + seconds
        
        if total_seconds <= 0:
            messagebox.showerror("Invalid Input", "Please enter a positive time.")
            return
        
        # Disable the start button while the timer is active
        start_button.config(state=tk.DISABLED)
        
        # Set the shutdown process time
        shutdown_process = time.time() + total_seconds
        
        # Hide the main window
        root.withdraw()
        
        # Start the countdown timer thread
        countdown_thread = threading.Thread(target=update_timer)
        countdown_thread.start()
        
        # Wait for the specified time before shutting down
        time.sleep(total_seconds)
        
        # Shut down the computer
        os.system("shutdown /s /f /t 0")
        
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter valid time values.")
    except KeyboardInterrupt:
        print("Timer was interrupted.")

def cancel_shutdown():
    global shutdown_process
    
    if shutdown_process is not None:
        os.system("shutdown /a")  # Cancel the shutdown
        shutdown_process = None
        timer_label.config(text="Time left: 00:00")
        start_button.config(state=tk.NORMAL)  # Re-enable the start button
        root.deiconify()  # Show the main window again
        messagebox.showinfo("Shutdown Canceled", "Shutdown has been canceled.")

# Create the main window
root = tk.Tk()
root.title("Shutdown Timer")



# Create and place widgets
minutes_label = tk.Label(root, text="Minutes:")
minutes_label.pack(pady=5)

minutes_entry = tk.Entry(root)
minutes_entry.pack(pady=5)

seconds_label = tk.Label(root, text="Seconds:")
seconds_label.pack(pady=5)

seconds_entry = tk.Entry(root)
seconds_entry.pack(pady=5)

start_button = tk.Button(root, text="Start Timer", command=shutdown_timer)
start_button.pack(pady=5)

cancel_button = tk.Button(root, text="Cancel Shutdown", command=cancel_shutdown)
cancel_button.pack(pady=10)

timer_label = tk.Label(root, text="Time left: 00:00")
timer_label.pack(pady=5)


# Run the GUI event loop
root.mainloop()
```
